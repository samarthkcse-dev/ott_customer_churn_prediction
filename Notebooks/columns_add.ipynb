```python
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/samar/Desktop/samarth15/OTT platform customer churn prediction/Data/Cleaned Data/users_cleaned.csv")
```

```python
df.head()
```

adding subcsription duration days column

```python
today=pd.Timestamp.today()
df["subscription_start_date"]=pd.to_datetime(df["subscription_start_date"])
df["subscription_duration_days"]=(today-df["subscription_start_date"]).dt.days
```

```python
df.head()
```

```python
df=df.drop("churn",axis=1)
```

Adding account_age_days Column

```python
df["created_at"].dtype
#df.info()
df["created_at"]=pd.to_datetime(df["created_at"])
df["account_age_days"]=(today-df["created_at"]).dt.days
```

adding senior_user column

```python
df["senior_user"] = (
    df["age"] >= 60
).astype(int)
```

```python
import numpy as np
```

Adding watch_hours column

```python
df["Watch_hours"]=np.random.randint(0,101,size=len(df))
```

Adding days_since_last_login column

```python
df["days_since_last_login"]=np.random.randint(0,61,size=len(df))
```

adding Auto renew column

```python
df["auto_renew"]=np.random.choice(["Yes","No"],size=len(df),p=[0.75,0.25])
```

adding payment failures column

```python
df["paymenr_failures"]=np.random.choice([0,1,2,3],size=len(df),p=[0.75,0.15,0.07,0.03])
```

```python
df.head()
```

```python
df.rename(columns={"Watch_hours":"watch_hours_L30_days"},inplace=True)
df.rename(columns={"paymenr_failures":"payment_failures"},inplace=True)
```

```python

conditions = (
    (df["watch_hours_L30_days"] < 15) &
    (df["days_since_last_login"] > 15) &
    (df["payment_failures"] >= 1) &
    (df["auto_renew"] == "No")
)

df["churn"] = np.where(conditions, 1, 0)
```

```python
df["Watch_hours"]=np.random.randint(0,61,size=len(df))
```

```python
df.to_csv("users_cleaned_final.csv",index=False)
```
