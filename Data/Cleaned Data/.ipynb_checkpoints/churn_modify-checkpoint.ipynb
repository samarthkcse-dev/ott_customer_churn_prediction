Re-assigning churn data

```python
import random

def assign_churn(days):
    if days > 45:
        return random.choices([1, 0], weights=[85, 15])[0]
    elif days > 30:
        return random.choices([1, 0], weights=[60, 40])[0]
    else:
        return random.choices([1, 0], weights=[10, 90])[0]

df["churn"] = df["days_since_last_login"].apply(assign_churn)
```
